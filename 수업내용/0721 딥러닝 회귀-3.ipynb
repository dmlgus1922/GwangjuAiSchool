```python
from sklearn.preprocessing import LabelEncoder
#모듈 로드
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential  #모델 만들기
from tensorflow.keras.layers import Dense, Dropout, Flatten  #레이어 쌓기
from tensorflow.keras.optimizers import SGD, RMSprop, Adam
from tensorflow.keras.losses import sparse_categorical_crossentropy, categorical_crossentropy, binary_crossentropy, mean_squared_error
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.models import load_model
```

```python
from tensorflow.keras.datasets import boston_housing
(X_data, y_data), (tt_x, tt_y)=boston_housing.load_data()
```

```python
X_data.shape
```

```python
y_data.shape
```

```python
np.unique(y_data) #실제값이 겁나 많으면 일반적으로 회귀문제
```

```python
m = Sequential()

m.add(Dense(26,activation='relu',input_shape=(13,)))
m.add(Dense(52, activation = 'relu'))
m.add(Dense(1)) #회귀 문제는 논리 연산이 필요한 분류 문제와 달리 회귀값만 나타나면 되므로 활성화함수가 필요 없다.
#분류 문제는 회귀값을 sigmoid, softmax 등의 출렴함수를 거쳐야 값이 나온다.
m.summary()
```

```python
m.compile(loss = 'mse', metrics=['mae']) #학습 방법: mse, 평가:mae(평균 절대값 오차)
```

```python
m.fit(X_data, y_data, epochs = 1000)
```
