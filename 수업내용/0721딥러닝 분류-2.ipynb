```python
import seaborn as sns
data = sns.load_dataset('iris')
data
```

```python
import numpy as np
np.unique(data['species'])
```

```python
from sklearn.preprocessing import LabelEncoder
#모듈 로드
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential  #모델 만들기
from tensorflow.keras.layers import Dense, Dropout, Flatten  #레이어 쌓기
from tensorflow.keras.optimizers import SGD, RMSprop, Adam
from tensorflow.keras.losses import sparse_categorical_crossentropy, categorical_crossentropy, binary_crossentropy
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.models import load_model
```

```python
X = data[['sepal_length','sepal_width','petal_length','petal_width']]
y = data['species']
X
```

```python
X.shape, y.shape
```

```python
l_m = LabelEncoder().fit(y)
l_m.classes_
```

```python
n_y = LabelEncoder().fit_transform(y)
```

```python
t_x, v_x, t_y, v_y = train_test_split(X,n_y, random_state = 42, test_size=0.2)
```

```python
m = Sequential()
m.add(Flatten(input_shape=(4,)))
m.add(Dense(100, activation='relu'))
m.add(Dense(3, activation='softmax'))
m.summary()
```

```python
m.compile(loss = 'sparse_categorical_crossentropy', metrics='accuracy')
```

```python
m.fit(t_x,t_y,epochs=100, validation_data=[v_x, v_y])
```

```python
k = m.predict(t_x[:10]) > 0.5
```

```python
for i in k:
    print(l_m.classes_[i])
```
